```python
FANPAGE_LINK ="losangeleslakers"
FOLDER_PATH = "/content/drive/MyDrive/python/"
COOKIE_PATH = "/content/drive/MyDrive/python/cookies.txt"
```

```python
from facebook_scraper import get_posts
import pandas as pd
import numpy as np
```

```python
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=20, cookies=COOKIE_PATH):
    post_list.append(post)
```

```python
post_df_full = pd.DataFrame(columns=post_list[0].keys(), index=range(len(post_list)), data = post_list)
excel_path = FOLDER_PATH + FANPAGE_LINK + ".xlsx"
post_df_full.to_excel(excel_path, index = False, encoding = 'utf-8')
print(excel_path)
```

```python
import numpy as np
numpy_path = FOLDER_PATH + FANPAGE_LINK + ".npy"
arr = np.array(post_list)
np.save(numpy_path, arr)
print(numpy_path)
```
